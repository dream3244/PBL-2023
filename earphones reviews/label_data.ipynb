## 耳机评价

```python
comments = pd.read_csv('comments.csv')
comments.rename(columns = {'Unnamed: 0': 'content_id', '0':'content'}, inplace = True)
comments.drop('Unnamed: 0.1', axis = 1, inplace = True)
```

### 关键词

```python
keywords = ['音质','价格','舒适度','外形','功能','配置', '其他']

function = ['续航','降噪','连接','蓝牙','快速','方便','电话','通话','驱动','电','可玩性'] # 功能

price = ['便宜','贵','有钱','低价','肾','低端','高端','顶级' ,'性价比', '成本','利润','价格','玩不起','亲民','太黑','有点黑','预算','差钱','缺钱']

shape = ['体积','臃肿','小巧','看着','外形','包装','质感','形状','盒子','袋','好看','难看','小不点','骚','漂亮','养眼','重量']

sound = ['音质','舒适','音染','音场','宽阔','曲线','电','噪音','木','金','质感','低音','高音','均衡','调和','声场','音色','冷','暖','厚实','圆润',
        '古典','钢琴','爵士','摇滚','鼓','吉他','电流','现场','3d','3D','立体','头场','流行','R&B','乡村','琴','动态','HIFI','hifi','Hifi','大气','恢弘','清丽','声音厚','煲机','温和',
        '粗糙','人声','毒','解析力','失真','好听','耐听','回声','声音','音量','低频','高频','中频','清晰','底噪','试听','音源','瑕疵']

comfort = ['入耳','舒适','疼','长时间','轰','舒服']

settings = ['数据','动圈','管子','线','耳塞','入耳','挂','电流','电','蓝牙', '焊','解码','折腾','没啥用','没用','水太深','指标','配置','单元','解析',
           '声道','低阻','阻抗','高阻','镁','铝','合金','铜']

others = []

def subject(keywords, comments):
    count = 0
    keyword_list = [sound, price, comfort, shape, function, settings, others]
    subject = 0
    for item in keyword_list[:-1]:
        for i in range(len(item)):
            if item[i] in comments:
                subject = keywords[keyword_list.index(item)]
            else: continue
    
    if subject == 0: subject = keywords[-1]
    return subject

subject = comments['content'].apply(lambda x:subject(keywords, x))
comments['subject'] = subject
```

### 情感词与情感分类

```python
good = ['干净','好','水准','牛','高级','舒适','好听','不错','强','追求','惊艳','好看','小巧','用心','精致','良心','充足', '不上当','good','nice','great','品质','正品','舒服','吊打','强',
       '猛','懂','值','赞','mark','Mark','种草','大气','恢弘','恢宏','骚','亲民','羡慕','恭喜','喜欢','温和','漂亮','有品位','有品味','哈哈','开心','高兴','养眼','支持','毒','厉害','可玩性',
       '耐听','合适','感人','厚实','圆润','显著','顶','难看','清晰','我选','优秀','出众','腻害','无损','惊喜']

bad = ['不耐听','不好','差','不喜欢','不方便','性价比低','轰','疼','不舒服','垃圾','辣鸡','lj','难听','混浊','模糊','噪音','音染','没有保证','无语','无奈','骗','上当','呵呵','难看','bad','假',
      '仿','一言难尽','玩不起','宽阔','折腾','没啥用','没用','没有高级感','水太深','粗糙','太黑','有点黑','哭','难过','菜','失真','扔掉','毁三观','送人','卖掉','没必要','扰','烦',
       '太贵','退','屎','急','受损'
      ]

sentiment_value = [-1, 0, 1]

def generate_sentiment_word(good, bad, comment):
    for item in bad:
        if item in comment: return item
    for item in good:
        if item in comment: return item
    return ''

def generate_sentiment_value(good, bad, comment):
    for item in bad:
        if item in comment: return -1
    for item in good:
        if item in comment: return 1
    return 0


sentiment_word  = comments['content'].apply(lambda x:generate_sentiment_word(good, bad, x))
comments['sentiment_word'] = sentiment_word

sentiment_value  = comments['content'].apply(lambda x:generate_sentiment_value(good, bad, x))
comments['sentiment_value'] = sentiment_value

comments
```

```python
comments.to_csv('erji_sentiments.csv')
```
